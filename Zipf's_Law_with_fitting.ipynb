```python
import numpy as np
fp = open('Alice.txt','r',encoding='UTF-8')

line = fp.readline()
my_dict = {}

while line:
    s = line.split()
    for x in s:
        if x not in my_dict:
            my_dict[x] = 1
        else:
            #my_dict[x] =  my_dict[x] + 1
            my_dict[x] += 1
    line = fp.readline()

fp.close()

num = []

for key in my_dict:
    num.append(my_dict[key])
    
print(len(num))
num.sort()
num.reverse()

import matplotlib.pyplot as plt

plt.loglog(range(len(num)),num)
plt.xlabel('Rank of the word')
plt.ylabel('Number of the occurance')
plt.show
```

```python
from scipy import optimize

def test_func(x, amp, alpha):
    return amp*(x**alpha)

x_data = np.linspace(1,len(num),num = len(num))
y_data = num
params, params_covariance = optimize.curve_fit(test_func, x_data, y_data,
                                               p0=[2, 2])


print('amp:',params[0])
print('alpha:',params[1])
print('covarience:',params_covariance)

plt.figure(figsize=(6, 4))
plt.loglog(range(len(num)),num)
plt.plot(x_data, test_func(x_data, params[0], params[1]),
         label='Fitted function')

plt.legend(loc='best')

plt.show()
```
